```
def fibonacci_naive(n: int) -> int:
    """
    Calculates the nth Fibonacci number using naive recursion.
    Time complexity: O(2^n) - very inefficient due to repeated calculations.
    """
    if n <= 1:
        return n

    # Repeatedly calculates fibonacci_naive(n-2) multiple times
    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2)

# print(f"F(6) Naive: {fibonacci_naive(6)}") # Output: 8
```
